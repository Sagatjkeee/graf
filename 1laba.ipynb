```python
#%%
#Xn+1 = (Xn + Xn-16) % 512
import random
import math
import matplotlib.pyplot as plt

i = 0
a = []
while i < 16:
    ch = (random.randrange(100))
    a.append(ch)
    i += 1
print(a)
x = 15
y = 0
i1 = 0
aa = []
while i1 < 484:
    b = ((a[x] + a[y]) % 512)
    #print(b)
    x += 1
    y += 1
    i1 += 1
    #print(b)
    aa.append(b)
    a.append(b)
aaa = []
t = 0
j = 0
while t < len(a):
    g = (a[j] / 512)
    j += 1
    t += 1
    aaa.append(g)
```

```python
m = math.fsum(aaa) / 500
print("Мат. ожидание равно: {}".format(m))
```

```python
result = 0
for i in range(len(aaa)):
    result += ((aaa[i] - m) **2)
d = result / 500
print("Дисперсия равна: {}".format(d))
```

```python
s = 2
bb = []
for num, chislo in enumerate(list(aaa)):
    if num + s < len(aaa):
        bb.append(chislo * aaa[num + s])
result = sum(bb)
r = 1 / 12 * result / (500 - s)
print("Коэфициент коррелляции: {}".format(r))
```

```python
kolvo = 500
b = [i for i in range(kolvo)]
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005
rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
plt.figure(figsize=(8, 8))


ax_scatter = plt.axes(rect_scatter)
ax_scatter.grid(which="major", linewidth=1.2, color="black")
ax_scatter.grid(which="minor", linestyle="--", color="green", linewidth=1)
ax_scatter.tick_params(which='major', length=10, width=2)
ax_histx = plt.axes(rect_histx)
ax_histx.grid(True)
ax_histx.tick_params(direction='in', labelbottom=False)
ax_scatter.scatter(aaa, b)
bins = 10
f1 = ax_histx.hist(aaa, bins, stacked=True)
ax_histx.set_xlim(ax_scatter.get_xlim())

plt.show()
```
